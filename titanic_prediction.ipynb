handling missing values

```python
#drop the cabin column
train=train.drop(columns='Cabin') #axis 0 for row, 1 for columns
```

```python
#replace age with the mean of all ages
#train['Age'].fillna(train['Age'].mean(), inplace=True) #inplace will save it in the dataset.
train['Age'] = train['Age'].fillna(train['Age'].mean())
```

```python
#for embarked, we'll find the most repeating value, aka mode
print(train['Embarked'].mode())
```

```python
print(train['Embarked'].mode()[0])
```

```python
#replacing the missing values in embarked column with S(MODE VALUE)
train['Embarked']=train['Embarked'].fillna(train['Embarked'].mode()[0], inplace=True)
```

seperating features and target

```python
X=train.drop(columns=['PassengerId', 'Name', 'Ticket','Survived'])
Y=train['Survived']
```

splitting data into training and testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2) #0.2 is 20%(test data)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

model training

logistic regression (binary classificaton)

```python
model= LogisticRegression()
```

```python
train.isnull().sum()
```

```python
print(X_train.isnull().sum())
```

```python
X_train['Age']=X_train['Age'].fillna(X_train['Age'].mean())
X_train['Embarked']=X_train['Embarked'].fillna(X_train['Embarked'].mode()[0])
```

```python
# Use SAME mean and mode (VERY IMPORTANT)

age_mean = X_train['Age'].mean()
embarked_mode = X_train['Embarked'].mode()[0]

X_train['Age'] = X_train['Age'].fillna(age_mean)
X_test['Age'] = X_test['Age'].fillna(age_mean)

X_train['Embarked'] = X_train['Embarked'].fillna(embarked_mode)
X_test['Embarked'] = X_test['Embarked'].fillna(embarked_mode)
```

```python
#training the LR model with training data(xtrain, ytrain)
model.fit(X_train, Y_train)
```
